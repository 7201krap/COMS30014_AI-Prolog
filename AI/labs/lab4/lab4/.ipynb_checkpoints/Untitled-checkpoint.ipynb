```python
import numpy as np
import re
import matplotlib.pyplot as plt

# initialisation and settings
max_ = []
min_ = []
mean_ = []

max_xs  = np.linspace(0, 1000, 1000)
min_xs  = np.linspace(0, 1000, 1000)
mean_xs = np.linspace(0, 1000, 1000)

f = open("ga_output.dat", 'r')
file_name = "ga_output.dat"
# initialisation and settings finishes here

def file_len(fname):
    with open(fname) as f:
        for i, l in enumerate(f):
            pass
    return i + 1


for i in range(file_len(file_name)):
    line = f.readline()
    line = re.findall("\d+\.\d+", line)

    # print(line)

    max_.append(line[0])
    min_.append(line[1])
    mean_.append(line[2])

f.close()


fig, ax = plt.subplots()
ax.plot(max_xs, max_, 'r')
ax.plot(min_xs, min_, 'g')
ax.plot(mean_xs, mean_, 'b')
# ax.set_xlim([0,1000])
# ax.set_ylim([0,1])
ax.set_xlabel('generation')
ax.set_ylabel('fitness')
plt.show()

```
